```python
from alpha_vantage.timeseries import TimeSeries
import requests
import pandas as pd
import io  # Import StringIO for handling CSV data
import os

# Replace 'YOUR_API_KEY' with your Alpha Vantage API key
api_key = os.getenv('ALPHA_VANTAGE_API_KEY')
```

# Intraday Data

```python
# Define parameters
symbol = "AAPL"
interval = "1min"
month = "2008-01"
extended_hours = "false"

# Construct the API URL
url = (
    f"https://www.alphavantage.co/query"
    f"?function=TIME_SERIES_INTRADAY"
    f"&symbol={symbol}"
    f"&interval={interval}"
    f"&month={month}"
    f"&extended_hours={extended_hours}"
    f"&apikey={api_key}"
    f"&datatype=csv"  # Request CSV output
)

# Fetch the data
response = requests.get(url)

# Check for successful response
if response.status_code == 200:
    # Parse the CSV response into a DataFrame
    data = pd.read_csv(io.StringIO(response.text))
    print(data.head())  # Display the first few rows
else:
    print(f"Error: {response.status_code}, {response.text}")
```

```python
# Display the data
print(data.head())
print(data.tail())
```

# News

```python
import os
import requests
import pandas as pd

def fetch_news_and_sentiment(api_key, tickers, topics, time_from, time_to, sort='LATEST', limit=1000):
    """
    Fetch news and sentiment data from Alpha Vantage API.

    Parameters:
    - api_key (str): Your Alpha Vantage API key.
    - tickers (str): Comma-separated list of tickers (e.g., 'AAPL,MSFT').
    - topics (str): Comma-separated list of topics (e.g., 'technology,finance').
    - time_from (str): Start date in the format 'YYYY-MM-DD'.
    - time_to (str): End date in the format 'YYYY-MM-DD'.
    - sort (str): Sort order of the results. Default is 'LATEST'.

    Returns:
    - pd.DataFrame: DataFrame containing news and sentiment data.
    """
    url = (
        f"https://www.alphavantage.co/query"
        f"?function=NEWS_SENTIMENT"
        f"&tickers={tickers}"
        # f"&topics={topics}"
        f"&time_from={time_from}T0000"
        f"&time_to={time_to}T0000"
        f"&sort={sort}"
        f"&apikey={api_key}"
        f"&limit={limit}"
    )

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if 'feed' in data:
            return pd.DataFrame(data['feed'])
        else:
            print("No news data found.")
            return pd.DataFrame()
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return pd.DataFrame()


# Example usage
api_key = os.getenv('ALPHA_VANTAGE_KEY')
tickers = 'AAPL'
topics = 'technology,finance'
time_from = '20220101'
time_to = '20220401'

data = fetch_news_and_sentiment(
    api_key, tickers, topics, time_from, time_to)
```
